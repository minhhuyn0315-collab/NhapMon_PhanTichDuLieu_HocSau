# Phân tích tần số và Loại bỏ Stopwords/Dấu câu

```python
import string

fd = nltk.FreqDist(macbeth)
print("10 từ thông dụng nhất:", fd.most_common(10))

sw = set(nltk.corpus.stopwords.words('english'))
punctuation = set(string.punctuation)

macbeth_filtered = [w.lower() for w in macbeth if w.lower() not in sw and w.lower() not in punctuation]

fd_filtered = nltk.FreqDist(macbeth_filtered)
print("10 từ thông dụng sau khi lọc:", fd_filtered.most_common(10))
```

# Rút trích từ theo đặc điểm

```python
long_words = [w for w in macbeth if len(w) > 12]
print("Từ dài > 12 ký tự:", sorted(long_words))

ious_words = set([w for w in macbeth if 'ious' in w])
print("Các từ chứa 'ious':", sorted(ious_words))
```

# Phân tích Bigrams và Trigrams

```python
bgrms = nltk.FreqDist(nltk.bigrams(macbeth_filtered))
print("15 Bigrams phổ biến:", bgrms.most_common(15))

tgrms = nltk.FreqDist(nltk.trigrams(macbeth_filtered))
print("10 Trigrams phổ biến:", tgrms.most_common(10))
```

# Phân tích cảm xúc (Sentiment Analysis) với Naïve Bayes

```python
import random

reviews = nltk.corpus.movie_reviews
documents = [(list(reviews.words(fileid)), category)
             for category in reviews.categories()
             for fileid in reviews.fileids(category)]
random.shuffle(documents)

all_words = nltk.FreqDist(w.lower() for w in reviews.words())
word_features = list(all_words)

def document_features(document, word_features):
    document_words = set(document)
    features = {}
    for word in word_features:
        features['contains({})'.format(word)] = (word in document_words)
    return features

featuresets = [(document_features(d, word_features), c) for (d, c) in documents]

train_set, test_set = featuresets[1500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)

print("Độ chính xác Naïve Bayes:", nltk.classify.accuracy(classifier, test_set))

classifier.show_most_informative_features(10)
```

```python
from nltk.corpus import stopwords

def check_stopword(word, lang):
    stop_words = set(stopwords.words(lang))
    if word.lower() in stop_words:
        return f"'{word}' LÀ một stopword trong tiếng {lang}."
    else:
        return f"'{word}' KHÔNG PHẢI là stopword trong tiếng {lang}."

print(check_stopword("the", "english"))
print(check_stopword("und", "german"))
print(check_stopword("apple", "english"))
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')

def remove_stopwords(text, lang='english'):
    stop_words = set(stopwords.words(lang))

    words = word_tokenize(text)

    filtered_text = [w for w in words if w.lower() not in stop_words]
    return filtered_text

text_sample = "AI is the future of technology and it will change the world."
result = remove_stopwords(text_sample)

print("Văn bản gốc:", text_sample)
print("Sau khi loại bỏ stopwords:", result)
```

```python
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

def lookup_wordnet(word):
    syns = wordnet.synsets(word)

    if not syns:
        return "Không tìm thấy dữ liệu cho từ này."

    print(f"Tra cứu từ: '{word}'")
    print(f"- Định nghĩa: {syns[0].definition()}")

    examples = syns[0].examples()
    if examples:
        print(f"- Ví dụ: {examples[0]}")
    else:
        print("- Ví dụ: Không có ví dụ sẵn có.")

lookup_wordnet("AI")
print("-" * 30)
lookup_wordnet("programming")
```

```python
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')

def get_synonyms_antonyms(word):
    synonyms = []
    antonyms = []

    for syn in wordnet.synsets(word):
        for l in syn.lemmas():
            synonyms.append(l.name())
            if l.antonyms():
                antonyms.append(l.antonyms()[0].name())

    print(f"Từ tra cứu: '{word}'")
    print(f"- Từ đồng nghĩa: {set(synonyms)}")
    print(f"- Từ trái nghĩa: {set(antonyms)}")

get_synonyms_antonyms("good")
```

```python
from nltk.corpus import wordnet

def compare_nouns(noun1, noun2):
    n1 = wordnet.synset(f'{noun1}.n.01')
    n2 = wordnet.synset(f'{noun2}.n.01')

    similarity = n1.path_similarity(n2)

    print(f"Độ tương đồng giữa '{noun1}' và '{noun2}': {similarity:.4f}")

compare_nouns("dog", "cat")
compare_nouns("dog", "car")
```

```python
import nltk
from nltk.corpus import wordnet

nltk.download('wordnet')
nltk.download('omw-1.4')

def compare_verbs(verb1, verb2):
    v1_synsets = wordnet.synsets(verb1, pos=wordnet.VERB)
    v2_synsets = wordnet.synsets(verb2, pos=wordnet.VERB)

    if v1_synsets and v2_synsets:
        similarity = v1_synsets[0].path_similarity(v2_synsets[0])
        print(f"Độ tương đồng giữa '{verb1}' và '{verb2}': {similarity:.4f}")
    else:
        print("Không tìm thấy dữ liệu cho một trong hai động từ.")

compare_verbs("run", "sprint")
compare_verbs("eat", "sleep")
```
